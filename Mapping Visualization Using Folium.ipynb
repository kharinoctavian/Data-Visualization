```python
#import data
df = pd.read_excel("data mapping fix.xlsx")

df.info()
```

```python
import pandas as pd
plant1= df[df['Plant']=='SJA 1']
plant1.tail(5)

plant2= df[df['Plant']=='SJA 2']
plant2.head(5)

plant3= df[df['Plant']=='SJA SMG']
plant3.head(5)
```

```python
cluster_1 = (-7.36521117731903, 112.66582507455959)
cluster_2 = (-6.39477302268097, 107.33131710419748)
cluster_smg = (-7.01763552268097, 110.35273133584919)

import folium
maps = folium.Map(
    location = [-6.200000, 106.816666],
    tiles='OpenStreetMap',
    zoom_start=10,
)
#ambil data dari file yang sudah dibenerin tambah latitude longitude
folium.Circle(location=cluster_1, popup='SJA 1', tooltip="SJA 1", color='red', radius=10000, fill=True, fill_color='red').add_to(maps)
folium.Circle(location=cluster_2, popup='SJA 2', tooltip="SJA 2", color='yellow', radius=10000, fill=True, fill_color='yellow').add_to(maps)
folium.Circle(location=cluster_smg, popup='SJA SMG', tooltip="SJA SMG", color='orange', radius=10000, fill=True, fill_color='orange').add_to(maps)
plant1.apply(lambda row: folium.Marker(location=[row['lat'],row['lng']], tooltip = str(row['Change']) +", "+str(row['Kota']) +", "+str(row['Plant']),
                                       popup=f"Lattitude:<br>{row['lat']}<br>"
                                       f"Longitude:<br>{row['lng']}<br>",
                                       icon=folium.map.Icon(color='purple', icon='industry', prefix='fa')).add_to(maps),axis =1)
plant2.apply(lambda row: folium.Marker(location=[row['lat'],row['lng']], tooltip = str(row['Change']) +", "+str(row['Kota']) +", "+str(row['Plant']),
                                       popup=f"Lattitude:<br>{row['lat']}<br>"
                                       f"Longitude:<br>{row['lng']}<br>",
                                       icon=folium.map.Icon(color='blue', icon='industry', prefix='fa')).add_to(maps),axis =1)
plant3.apply(lambda row: folium.Marker(location=[row['lat'],row['lng']], tooltip = str(row['Change']) +", "+str(row['Kota']) +", "+str(row['Plant']),
                                       popup=f"Lattitude:<br>{row['lat']}<br>"
                                       f"Longitude:<br>{row['lng']}<br>",
                                       icon=folium.map.Icon(color='green', icon='industry', prefix='fa')).add_to(maps),axis =1)
maps
```
